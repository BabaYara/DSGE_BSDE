# KFAC PINN Example
This notebook demonstrates the new `KFACPINNSolver` and the simple `kfac_update` routine.

```python
import jax.numpy as jnp
from bsde_dsgE.kfac import KFACPINNSolver

# Dummy network and loss function
net = lambda x: x
loss_fn = lambda net, x: jnp.sum(net(x)**2)

solver = KFACPINNSolver(net=net, loss_fn=loss_fn, num_steps=3)
losses = solver.run(jnp.array([0.0]), jnp.zeros(2, dtype=jnp.uint32))
print(losses)
```
