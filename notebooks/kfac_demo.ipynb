# KFAC PINN Demo
This notebook demonstrates the `KFACPINNSolver`.

## Step-by-step

1. Create a network with Equinox.
2. Define the residual or loss function.
3. Instantiate `KFACPINNSolver`.
4. Call `solver.run` to train.
5. Inspect the loss history.

```python
import jax.numpy as jnp
from bsde_dsgE.optim import KFACPINNSolver

# Dummy network and loss function
net = lambda x: x
loss_fn = lambda net, x: jnp.sum(net(x)**2)

solver = KFACPINNSolver(net=net, loss_fn=loss_fn, num_steps=3)
losses = solver.run(jnp.array([0.0]), jnp.zeros(2, dtype=jnp.uint32))
print(losses)
```
